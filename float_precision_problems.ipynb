```python
import warnings as w
from numpy import *

w.filterwarnings("ignore")
_ = seterr(all="ignore")
```

```python
f = lambda x: f"{x:9.3e}"
R = []

a = 1e-8
R.append(("Small Numbers", f(float64(a)), f(float16(a)), "1e-8"))

a = 70000.0
R.append(("Big Numbers", f(float64(a)), f(float16(a)), "70000.0"))

n = int(1e5)
a = full(n, 1e-1)
R.append(("Long Sum", f(a.sum()), f(a.astype(float16).sum()),
          "sum([0.1 for k in range(int(1e5))])"))

a = 1.0
b = 1e-4
c = float64(a) - (float64(a) - float64(b))
d = float16(a) - (float16(a) - float16(b))
R.append(("Catastrophic Cancellation", f(c), f(d),
          "1.0 - (1.0 - 1e-4)"))

H = ("Description", "float64", "float16", "Symbolic Term")
W = [max(len(H[i]), max(len(r[i]) for r in R)) for i in range(4)]

def F(c, A=("l", "r", "r", "l")):
    return "  ".join(
        (c[i].rjust if A[i] == "r" else c[i].ljust)(W[i]) for i in range(4)
    )
```

```python
print(F(H))
print("  ".join("-" * x for x in W))
for r in R:
    print(F(r))
```
